```python
import numpy as np # vetor
import pandas as pd # banco de dados
import matplotlib.pyplot as plt # gráficos
```

# Passo 1 – Criar o dicionário principal

```python
distribuicoes = {'Normal': {'params': {'loc': 0, 'scale':1}, 'dados':None, 'estatisticas': None},
                 'Exponencial': {'params': {'scale':2}, 'dados':None, 'estatisticas':None},
                 'Uniforme': {'params':{'low': -1, 'high':1}, 'dados': None, 'estatisticas':None}}
```

# Passo 2 – Importar bibliotecas


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Passo 3 – Gerar os dados

```python
normal = np.random.normal(0, 1, size = 1000)
exp = np.random.exponential(2, size = 1000)
unif = np.random.uniform(-1, 1, size = 1000)
```

```python
distribuicoes['Normal']['dados']= normal
distribuicoes['Exponencial']['dados']=exp
distribuicoes['Uniforme']['dados']=unif
```

```python
distribuicoes
```

# Passo 4 – Calcular estatísticas básicas

```python
distribuicoes['Normal']['estatisticas'] = {'média': np.mean(normal), 'mediana': np.median(normal), 'variância': np.var(normal), 'desvio padrão': np.std(normal)}
distribuicoes['Exponencial']['estatisticas'] = {'média': np.mean(exp), 'mediana': np.median(exp), 'variância': np.var(exp), 'desvio padrão': np.std(exp)}
distribuicoes['Uniforme']['estatisticas'] =  {'média': np.mean(unif), 'mediana': np.median(unif), 'variância': np.var(unif), 'desvio padrão': np.std(unif)}
```

# Passo 5 – Visualizar os dados

```python
plt.hist(normal)
```

```python
plt.hist(exp)
```

```python
plt.hist(unif)
```

# Passo 6 – Organizar resultados em Pandas

```python
estat = {'Normal': distribuicoes['Normal']['estatisticas'],
         'Exponencial': distribuicoes['Exponencial']['estatisticas'],
         'Uniforme': distribuicoes['Uniforme']['estatisticas']}
```

```python
pd.DataFrame(estat)
```

# Passo 7 – Adicionar novas distribuições


```python
distribuicoes['Poisson'] = {'params': {'lam': 2}, 'dados': None, 'estatisticas': None}
distribuicoes['Binomial'] = {'params': {'n': 10, 'p': 0.28}, 'dados': None, 'estatisticas': None}
```

```python
poisson = np.random.poisson(2, size = 1000)
binomial = np.random.binomial(10, 0.2, size = 1000)
```

```python
distribuicoes['Poisson']['dados'] = poisson
distribuicoes['Binomial']['dados'] = binomial
```

```python
distribuicoes['Poisson']['estatisticas'] = {'média': np.mean(poisson), 'mediana': np.median(poisson), 'variância': np.var(poisson), 'desvio padrão': np.std(poisson)}
distribuicoes['Binomial']['estatisticas'] = {'média': np.mean(binomial), 'mediana': np.median(binomial), 'variância': np.var(binomial), 'desvio padrão': np.std(binomial)}
```

```python
estat['Poisson'] = distribuicoes['Poisson']['estatisticas']
estat['Binomial'] = distribuicoes['Binomial']['estatisticas']
```

```python
pd.DataFrame(estat)
```

# Passo 8 – Estatísticas adicionais

```python
distribuicoes['Normal']['estatisticas'].update({'mínimo': np.min(normal), 'máximo': np.max(normal), 'Q1': np.percentile(normal, 25), 'Q3': np.percentile(normal, 75)})
distribuicoes['Exponencial']['estatisticas'].update({'mínimo': np.min(exp), 'máximo': np.max(exp), 'Q1': np.percentile(exp, 25), 'Q3': np.percentile(exp, 75)})
distribuicoes['Uniforme']['estatisticas'].update({'mínimo': np.min(unif), 'máximo': np.max(unif), 'Q1': np.percentile(unif, 25), 'Q3': np.percentile(unif, 75)})
distribuicoes['Poisson']['estatisticas'].update({'mínimo': np.min(poisson), 'máximo': np.max(poisson), 'Q1': np.percentile(poisson, 25), 'Q3': np.percentile(poisson, 75)})
distribuicoes['Binomial']['estatisticas'].update({'mínimo': np.min(binomial), 'máximo': np.max(binomial), 'Q1': np.percentile(binomial, 25), 'Q3': np.percentile(binomial, 75)})
```

```python
estat = {'Normal': distribuicoes['Normal']['estatisticas'],
         'Exponencial': distribuicoes['Exponencial']['estatisticas'],
         'Uniforme': distribuicoes['Uniforme']['estatisticas'],
         'Poisson': distribuicoes['Poisson']['estatisticas'],
         'Binomial': distribuicoes['Binomial']['estatisticas']}

pd.DataFrame(estat)
```

# Passo 9 – Comparar distribuições graficamente

```python
plt.boxplot([normal, exp, unif, poisson, binomial],labels=['Normal', 'Exponencial', 'Uniforme', 'Poisson', 'Binomial'])
plt.title('Boxplot das Distribuições')
plt.xlabel('Distribuições')
plt.ylabel('Valores')

plt.show()
```
