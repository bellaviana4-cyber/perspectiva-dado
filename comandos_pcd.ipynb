# COMANDOS APRENDIDOS
## TERMINAL
1. `ctrl alt t`: abrir o terminal no linux
2. `ctrl shift v`: copiar e colar
3. `ls`: lista todos os arquivos da pasta
4. `ls -a`: mostra todos os arquivos da ṕasta inclusive os arquivos ocultos 
5. `cd` - muda para a pasta selecionada
6. `cd ..` - volta para os arquivos da past
7. `cat {arquivo}`: mostra o que tem no arquivo
8. `vi {arquivo}`: para editar/modificar o arquivo
9. 


## GIT
1. `git clone`: clonar o repositório na máquina
2. `git status`: verificar o status do arquivo, ou seja, se há modificações nele 
3. `git diff`: verificar as alterações do arquivo
4. `git add .`: adiciona todas alterações feitas
5. `git add {nome do arquivo}`: adiciona as alterações feitas no arquivo especificado
6. `git commit -m 'mensagem'`: comentar as mudanças feitas no git add
7. `git push`: envia os arquivos comitados ao repositório
8. `git pull`: sincroniza as alterações feitas no repositório na máquina 
9. `git restore --staged <file>`: para mudar algo no add antes do commit
10. `touch .gitignore`: cria arquivo .gitignore 

## UV
1. wget -qO- https://astral.sh/uv/install.sh | sh: para instalar o uv na máquina
2. `uv init`: para inicializar o uv
3. `uv add {pacote}`: para instalar pacotes nas dependências
4. `uv run jupyter notebook`: cria notebook

## VI
1. `esc u`: desfaz a última ação
2. `i`: entrar no modo edição
3. `o`: pular uma linha e entrar no modo edição
4. `esc`: sair do modo de edição (voltar ao modo visualização)
5. `:q!`: sair sem salvar as alterações
6. `:w` : salvar as alterações
7. `:q` : sair do modo vi
8. `x`: apaga o caractere
9. `dd`: deleta a linha completa
